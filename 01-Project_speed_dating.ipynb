# Speed Dating Data Key Statistics

```python
# gender : Female = 0 , Male = 1
dataset['gender'] = dataset['gender'].map({ 0 : "Female", 1 : "Male"})
# race of partner :
#   Black/African American=1
#	European/Caucasian-American=2
#	Latino/Hispanic American=3
#	Asian/Pacific Islander/Asian-American=4
#	Native American=5
#	Other=6
dataset['race'] = dataset['race'].map({ 1 : "Black/African",
                                        2 : "European/Caucasian",
                                        3 : "Latino/Hispanic",
                                        4 : "Asian/Pacific Islander/Asian",
                                        5 : "Native",
                                        6 : "Other"})
```

```python
print('number of participants : ', len(dataset['iid'].unique()))
print('number of waves : ', len(dataset['wave'].unique()))
```

```python
pd.set_option('display.max_rows',10)
df_ind = dataset.groupby('iid')['gender', 'age', 'race'].head(1)
```

```python
# partners stated preference at Time 1 for all 6 attributes
preference = dataset[['pf_o_att', 'pf_o_sin','pf_o_int','pf_o_fun', 'pf_o_amb', 'pf_o_sha']].mean()
labels = ['Attractive', 'Sincere', 'Intelligent', 'Fun', 'Ambitious', 'Shared interests']
plt.figure(figsize=(8,8))
plt.pie(preference ,labels = labels, autopct='%1.1f%%')
plt.title("partners stated preference")
plt.show()
```

```python
# look for in the opposite sex
look = dataset[['attr1_1','sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']].mean()
labels = ['Attractive', 'Sincere', 'Intelligent', 'Fun', 'Ambitious', 'Shared interests']
plt.figure(figsize=(8,8))
plt.pie(look ,labels = labels, autopct='%1.1f%%')
plt.title(" look for in the opposite sex")
plt.show()
```

```python
# partners stated preference at Time 1 for all 6 attributes
look2 = dataset[['attr7_2', 'sinc7_2', 'intel7_2', 'fun7_2', 'amb7_2', 'shar7_2']].mean()
labels = ['Attractive', 'Sincere', 'Intelligent', 'Fun', 'Ambitious', 'Shared interests']
plt.figure(figsize=(8,8))
plt.pie(look2 ,labels = labels, autopct='%1.1f%%')
plt.title(" best reflects the actual importance of these attributes in your decisions")
plt.show()
```

```python
list_rating = ['attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o']
rating = dataset.groupby('dec_o')[list_rating].mean().reset_index()
rating
```

```python
data = pd.pivot_table(rating, values = ['attr_o','sinc_o','intel_o','fun_o','amb_o','shar_o'], columns = 'dec_o')
data
```

```python
# rating by partner the night of the event, for all 6 attributes
labels = [ 'Attractive', 'Sincere', 'Intelligent', 'Fun', 'Ambitious', 'Shared interests']
plt.figure(figsize=(12,12))

plt.subplot(221)
plt.bar(height = data.iloc[:,0], x = labels )
plt.title("rating by partner the night of the event, for all 6 attributes")
plt.xlabel("decision of partner : No")
plt.subplot(222)
plt.bar(height = data.iloc[:,1], x = labels, color ='orange' )
plt.title("rating by partner the night of the event, for all 6 attributes")
plt.xlabel("decision of partner : Yes")
plt.show()
```

```python
list_activities = ['sports', 'tvsports', 'exercise', 'dining', 'museums','art', 'hiking',
	'gaming', 'clubbing', 'reading', 'tv','theater', 'movies', 'concerts', 'music', 'shopping', 'yoga']
df_activities = dataset.groupby('match')[list_activities].mean().reset_index()
df_activities
```

```python
data = pd.pivot_table(df_activities, values = list_activities, columns = 'match')
data
```

```python
#  How interested are you in the following activities, on a scale of 1-10?
labels = list_activities

plt.figure(figsize=(12,12))

plt.subplot(221)
plt.bar(height = data.iloc[:,0], x = labels )
plt.title("the following activities, on a scale of 1-10, match : no")
plt.xlabel("decision of partner : No")
plt.subplot(222)
plt.bar(height = data.iloc[:,1], x = labels, color ='orange' )
plt.title("the following activities, on a scale of 1-10, match : yes")
plt.xlabel("decision of partner : Yes")
plt.show()
```
